This project explored three different modeling approaches to attempt to find a profitable betting strategy on the over/under market in the UK Premier League. The models were:

- Random Forest: predicting the binary outcome of high scoring or not.
- Poisson model based on average goals scored and conceded per game.
- Random Forest model based on continuous output of expected payout from a bet on high scoring.

Key takeaways:
- All models were of some use.
- Both ML models had a precision greater than random chance (of games identified as high scoring, 0.55 and 0.57 respectively).
- The most profitable approach was to use the expected payout to train the data, suggesting the model is better at finding mispriced bets than outpredicting the bookies.
- We were able to record a small profit using this method, but it suffers from a small data set, overly simplistic model, and missing information.
- A real production-grade model could be developed based on this approach, but it would require much extended model inputs and greater model adaptability.
